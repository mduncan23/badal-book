## Train/Test Split

```python
# Create a custom function to split a train/test set
def shuffle_and_split_data(data, test_ratio, random_state=42):
    
    '''
    Input a dataframe and the ratio of test data.
    
    Returns split dataframe: Train_set, Test_set
    '''
    # set random state seed
    np.random.seed(random_state)
    # Shuffle the index values
    shuffled_indices = np.random.permutation(len(data))
    # Get a ratio of the total length of dataset
    test_set_size = int(len(data) * test_ratio)
    # slice the shuffled index for test data
    test_indices = shuffled_indices[:test_set_size]
    # slice the shuffled index for train data
    train_indices = shuffled_indices[test_set_size:]
    # return sliced train data and test data based on index
    return data.iloc[train_indices], data.iloc[test_indices]
```

Issues: random state will vary if the dataset changes. We need to avoid this and so will implement a unique identifier system.

**compute a hash of each instances identifier and put that instance in the test set if the hash is lower than or equal to 20% of the maximum hash value**

```python
# create function to see if id is in test set
def is_id_in_test_set(identifier, test_ratio):
    return crc32(np.int64(identifier)) < test_ratio * 2**32

# create function to split data based on unique ID
def split_data_with_id_hash(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: is_id_in_test_set(id_, test_ratio))
    # tilde (~) basically acts as 'not in'
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
# No identifier in data, reset index to create identifier
# this is only helpful if new data is always appended to the bottom of the dataframe
housing_with_id = housing.reset_index()
housing_with_id.head(3)
```

## Feature Engineering

Need to complete a train test split but ensure that income is split evenly, though still randomly

```python
# median income is important to the model
housing['median_income'].hist(bins='auto');
```

```python
# need to bin it to ensure an even split
housing['income_cat'] = pd.cut(housing['median_income'],
                               bins=[0, 1.5, 3, 4.5, 6, np.inf],
                               labels=[1, 2, 3, 4, 5])
```

### New Train/Test Split

```python
# instantiate a splitter
splitter = StratifiedShuffleSplit(n_splits=10,
                                  test_size=.2,
                                  random_state=42)
# create empty list to store values
strat_splits = []

# looping through the folds, appending the splits to the list
for train_index, test_index in splitter.split(housing, housing['income_cat']):
    # train data
    strat_train_set_n = housing.iloc[train_index]
    # test data
    strat_test_set_n = housing.iloc[test_index]
    strat_splits.append([strat_train_set_n, strat_test_set_n])
```

```python
# confirming that train split was done evenly
strat_splits[0][0]['income_cat'].value_counts(normalize=True)
```

```python
# confirming that test split was done evenly
strat_splits[0][1]['income_cat'].value_counts(normalize=True)
```

```python
# saving first split as train/test sets
strat_train_set, strat_test_set = strat_splits[0]
```

## Custom Classes

```python
# Creating a custom transformer that is similar to Standard Scaler
class StandardScalerClone(BaseEstimator, TransformerMixin):
    def __init__(self, with_mean=True): # no *args or **kwargs!
        self.with_mean = with_mean
        
    def fit(self, X, y=None): # y is required even though we dont use it
        X = check_array(X) # checks that X is an array with finite floar values
        self.mean_ = X.mean(axis=0)
        self.scale_ = X.std(axis=0)
        self.n_features_in_ = X.shape[1] # every estimator stores this in fit()
        return self # always return self!
    
    def transform(self, X):
        check_is_fitted(self) # looks for learned attributes (with trailing _)
        X = check_array(X)
        assert self.n_features_in_ == X.shape[1]
        if self.with_mean:
            X = X - self.mean_
        return X / self.scale_
```

```python
# Creating a custom transformer that is similar to Standard Scaler
class StandardScalerClone(BaseEstimator, TransformerMixin):
    """Custom transformer class that scales features to have zero mean and unit variance."""

    def __init__(self, with_mean: bool = True):
        """
        Initialize the StandardScalerClone class.

        Parameters:
            with_mean (bool): Whether to center the data or not (default=True).
        """
        self.with_mean = with_mean

    def fit(self, X, y=None):
        """
        Compute the mean and standard deviation of each feature of X.

        Parameters:
            X (ndarray): Array of shape (n_samples, n_features) containing the training data.
            y (ndarray): Array of shape (n_samples,) containing the target values.

        Returns:
            self (object): Returns the instance itself.
        """
        X = check_array(X)  # checks that X is an array with finite float values
        self.mean_ = X.mean(axis=0)
        self.scale_ = X.std(axis=0)
        self.n_features_in_ = X.shape[1]  # every estimator stores this in fit()
        return self

    def transform(self, X):
        """
        Scale the input data.

        Parameters:
            X (ndarray): Array of shape (n_samples, n_features) containing the input data.

        Returns:
            X_scaled (ndarray): Array of shape (n_samples, n_features) containing the scaled data.
        """
        check_is_fitted(self)  # looks for learned attributes (with trailing _)
        X = check_array(X)
        assert self.n_features_in_ == X.shape[1]
        if self.with_mean:
            X = X - self.mean_
        X_scaled = X / self.scale_
        return X_scaled
```

### Kmeans Cluster similarity with RBF

```python
class ClusterSimilarity(BaseEstimator, TransformerMixin):
    def __init__(self, n_clusters=10, gamma=1.0, random_state=None):
        self.n_clusters = n_clusters
        self.gamma = gamma
        self.random_state = random_state
        
    def fit(self, X, y=None, sample_weight=None):
        self.kmeans_ = KMeans(self.n_clusters, random_state=self.random_state)
        self.kmeans_.fit(X, sample_weight=sample_weight)
        return self # always return self!
    
    def transform(self, X):
        return rbf_kernel(X, self.kmeans_.cluster_centers_, gamma=self.gamma)
    
    def get_feature_names_out(self, names=None):
        return [f'Cluster {i} similarity' for i in range(self.n_clusters)]
```

```python
from typing import Optional, List

class ClusterSimilarity2(BaseEstimator, TransformerMixin):
    """Transforms input data by computing RBF kernel similarity between each sample and KMeans cluster centers."""
    
    def __init__(self, n_clusters: int = 10, gamma: float = 1.0, random_state: Optional[int] = None):
        """
        Initializes ClusterSimilarity transformer.
        
        Parameters:
            n_clusters (int): Number of clusters to form.
            gamma (float): Kernel coefficient for 'rbf' kernel.
            random_state (int): Seed used by the random number generator.
        """
        self.n_clusters = n_clusters
        self.gamma = gamma
        self.random_state = random_state
        
    def fit(self, X: np.ndarray, y: Optional[np.ndarray] = None, sample_weight: Optional[np.ndarray] = None) -> 'ClusterSimilarity':
        """
        Computes KMeans clustering on input data.
        
        Parameters:
            X (np.ndarray): Input data.
            y (Optional[np.ndarray]): Ignored parameter.
            sample_weight (Optional[np.ndarray]): Individual weights for each sample.
        
        Returns:
            self: The fitted transformer.
        """
        self.kmeans_ = KMeans(self.n_clusters, random_state=self.random_state)
        self.kmeans_.fit(X, sample_weight=sample_weight)
        return self # always return self!
    
    def transform(self, X: np.ndarray) -> np.ndarray:
        """
        Computes RBF kernel similarity between input data and KMeans cluster centers.
        
        Parameters:
            X (np.ndarray): Input data.
        
        Returns:
            np.ndarray: Similarity matrix.
        """
        return rbf_kernel(X, self.kmeans_.cluster_centers_, gamma=self.gamma)
    
    def get_feature_names_out(self, names: Optional[List[str]] = None) -> List[str]:
        """
        Returns the names of the output features.
        
        Parameters:
            names (Optional[List[str]]): Ignored parameter.
            
        Returns:
            List[str]: Output feature names.
        """
        return [f'Cluster {i} similarity' for i in range(self.n_clusters)]
```

```python
def column_ratio(X):
    """
    Returns the ratio of the first column to the second column of the input array.

    Args:
    - X: numpy array of shape (n_samples, 2)

    Returns:
    - numpy array of shape (n_samples, 1)
    """
    return X[:, [0]] / X[:, [1]]


def ratio_name(function_transformer, feature_names_in):
    """
    Returns a list containing the name 'ratio'.

    Args:
    - function_transformer: function transformer object
    - feature_names_in: list of input feature names

    Returns:
    - list containing the name 'ratio'
    """
    return ['ratio']


def ratio_pipeline():
    """
    Returns a pipeline object for calculating column ratios and scaling the data.

    Returns:
    - pipeline object
    """
    return make_pipeline(
        SimpleImputer(strategy='median'),
        FunctionTransformer(column_ratio, feature_names_out=ratio_name),
        StandardScaler()
    )


log_pipeline = make_pipeline(
    SimpleImputer(strategy='median'),
    FunctionTransformer(np.log, feature_names_out='one-to-one'),
    StandardScaler()
)


cluster_simil = ClusterSimilarity(n_clusters=10, gamma=1, random_state=42)


default_num_pipeline = make_pipeline(
    SimpleImputer(strategy='median'),
    StandardScaler()
)


preprocessing = ColumnTransformer([
    ('bedrooms', ratio_pipeline(), ['total_bedrooms', 'total_rooms']),
    ('rooms_per_house', ratio_pipeline(), ['total_rooms', 'households']),
    ('people_per_house', ratio_pipeline(), ['population', 'households']),
    ('log', log_pipeline, ['total_bedrooms', 'total_rooms', 'population',
                           'households', 'median_income']),
    ('geo', cluster_simil, ['latitude', 'longitude']),
    ('cat', cat_pipeline, make_column_selector(dtype_include=object))    
],
remainder=default_num_pipeline) # one column remaining: housing_median_age
```
